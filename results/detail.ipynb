```python
import seaborn as sns
import pandas

sns.set(style="ticks")
sns.set_style("darkgrid")
sns.set(rc={"figure.figsize":(12, 10)})
```

```python
data = pandas.read_csv("out/detail.csv")

dfm = data[["num", "op", "avg", "std"]].melt(
    ["num", "op"], var_name="measurements", value_name="time")
ax = sns.relplot(kind="line", data=dfm, x="num", y="time", hue="measurements", row="op", height=10)
# ax.set_yscale("log")
ax.set(ylim=(0, 250))
ax.set(ylabel="time in ns")
```

```python
data = pandas.read_csv("out/detail.csv")

dfm = data[data["op"] == "get"]
dfm = data[["num", "avg", "std", "min", "max"]].melt(
    "num", var_name="measurements", value_name="time")
ax = sns.lineplot(data=dfm, x="num", y="time", hue="measurements")
ax.set_yscale("log")
ax.set_ylabel("time in ns")
```
